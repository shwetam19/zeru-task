```python
# Import the necessary libraries
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for plots
sns.set_style("whitegrid")

# --- Load the JSON file into a pandas DataFrame ---
try:
    with open('user-wallet-transactions.json', 'r') as f:
        data = json.load(f)
    df = pd.DataFrame(data)
    print("✅ Data loaded successfully!")
except FileNotFoundError:
    print("❌ Error: 'user-wallet-transactions.json' not found.")

# --- Display the first 5 rows to understand the structure ---
print("\n--- 1. First 5 Transactions ---")
display(df.head())

# --- Get a concise summary of the DataFrame ---
print("\n--- 2. Dataframe Info ---")
df.info()

# --- Count and visualize the occurrences of each transaction type ---
print("\n--- 3. Transaction Counts ---")
action_counts = df['action'].value_counts()
print(action_counts)

# Create a bar plot to visualize the counts
plt.figure(figsize=(10, 6))
sns.barplot(x=action_counts.index, y=action_counts.values, palette="viridis")
plt.title('Distribution of Transaction Actions', fontsize=16)
plt.ylabel('Number of Transactions', fontsize=12)
plt.show()

# Make sure you have already run the first cell and have the DataFrame 'df' loaded.

# ==============================================================================
# PHASE 2: FEATURE ENGINEERING
# ==============================================================================
print("--- Running Phase 2: Feature Engineering ---")

# Pre-processing
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df['amount'] = df['actionData'].apply(lambda x: float(x.get('amount', 0)))

# Aggregate Data by Wallet
wallet_features = df.groupby('userWallet').agg(
    wallet_age_days=('timestamp', lambda date: (date.max() - date.min()).days),
    transaction_count=('timestamp', 'count')
)

# Calculate Action-Specific Metrics using a pivot table
action_pivot = df.pivot_table(
    index='userWallet',
    columns='action',
    values='amount',
    aggfunc=['sum', 'count']
).fillna(0)
action_pivot.columns = ['_'.join(col) for col in action_pivot.columns]

# Merge all features
wallet_features = wallet_features.merge(action_pivot, on='userWallet', how='left').fillna(0)

# Engineer High-Level Behavioral Ratios & Risk Flags
total_borrow_volume = wallet_features.get('sum_borrow', 0)
wallet_features['repay_to_borrow_ratio'] = wallet_features.get('sum_repay', 0) / (total_borrow_volume + 1e-6)
wallet_features['deposit_to_borrow_ratio'] = wallet_features.get('sum_deposit', 0) / (total_borrow_volume + 1e-6)
wallet_features['liquidation_count'] = wallet_features.get('count_liquidationcall', 0)
wallet_features['is_liquidated'] = (wallet_features['liquidation_count'] > 0).astype(int)

print("✅ Phase 2 Complete. `wallet_features` DataFrame is created.\n")

# ==============================================================================
# PHASE 3: MODEL BUILDING AND SCORING
# ==============================================================================
print("--- Running Phase 3: Model Building and Scoring ---")
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Select features for clustering
features_for_model = [
    'wallet_age_days', 'transaction_count', 'repay_to_borrow_ratio',
    'deposit_to_borrow_ratio', 'is_liquidated', 'sum_borrow'
]
model_data = wallet_features[features_for_model].copy()

# Scale features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(model_data)

# Run K-Means
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
wallet_features['cluster'] = kmeans.fit_predict(scaled_features)

# Analyze and score clusters
cluster_analysis = wallet_features.groupby('cluster')[features_for_model].mean()
cluster_analysis['risk_rank'] = (
    cluster_analysis['is_liquidated'] * 100 -
    cluster_analysis['repay_to_borrow_ratio'] * 10 +
    np.log1p(cluster_analysis['sum_borrow']) / 100
)
sorted_clusters = cluster_analysis['risk_rank'].sort_values().index
score_mapping = {cluster_id: score for cluster_id, score in zip(sorted_clusters, np.linspace(950, 50, n_clusters).astype(int))}

# Assign final credit score
wallet_features['credit_score'] = wallet_features['cluster'].map(score_mapping)
print("✅ Phase 3 Complete. Credit scores are assigned.\n")


# ==============================================================================
# PHASE 4: FINAL DELIVERABLES
# ==============================================================================
print("--- Running Phase 4: Final Deliverables ---")
import matplotlib.pyplot as plt
import seaborn as sns

# Generate and Save the Score Distribution Graph
plt.figure(figsize=(10, 6))
sns.histplot(wallet_features['credit_score'], bins=10, kde=False, color='#4A90E2')
plt.title('Credit Score Distribution', fontsize=16)
plt.xlabel('Credit Score Range', fontsize=12)
plt.ylabel('Number of Wallets', fontsize=12)
plt.xticks(np.arange(0, 1001, 100))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('score_distribution.png')
print("📊 Graph saved as 'score_distribution.png'")

# Save final scores to JSON file
wallet_features[['credit_score']].to_json('wallet_scores.json', orient='index', indent=4)
print("📄 Final scores saved to 'wallet_scores.json'")
print("\n🎉 All steps are complete!")
```
